```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Dataset_w6 - Data.csv')
```

```python
df.head()
```

# Find out how many countries are there in the dataset.

```python
df['country'].count()
```

# Examine the data by using visualizations.

```python
#Count plot of region
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x=df['Region'])
plt.title('Number of countries by region',fontweight='bold',fontsize=20)
plt.xlabel('Region',fontsize=15)
plt.ylabel('Number',fontsize=15)
plt.xticks(rotation=90)
plt.yticks()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='country',y='GDP per capita (current US$)',data=df.sort_values('GDP per capita (current US$)',ascending=False).head(10))
plt.title('Top 10 countries based on GDP per capita (current US$)',fontweight='bold',fontsize=15)
plt.xlabel('Country',fontsize=15)
plt.ylabel('GDP per capita (current US$)',fontsize=15)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
sns.pairplot(df)
```

# Find out how many countries with respect to the region are available in the dataset



```python
df.groupby('Region').agg(countries_wrt_region=('country','count'))
```

# On average which region has the highest GDP per capita (current US$)

```python
highest_GDP_average=df.groupby("Region")[["GDP per capita (current US$)"]].mean()
highest_GDP_average.sort_values(by="GDP per capita (current US$)",ascending=False)
```

# On average which region has the International trade: Imports (million US$)


```python
highest_int_average=df.groupby("Region")[["International trade: Imports (million US$)"]].mean()
highest_int_average.sort_values(by="International trade: Imports (million US$)",ascending=False)
```

# Find out which region has more consistent with respect to GDP per capita (current US$


```python
consistent_GDP=df.groupby('Region')[['GDP per capita (current US$)']].std()
```

```python
consistent_GDP.sort_values(by='GDP per capita (current US$)',ascending=True)
```
