```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Telco-Customer-Churn.xlsx - Telco-Customer-Churn.csv')
```

```python
df.head()
```

# Compare churn count with respect to gender.

```python
plt.figure(figsize=(10,10))
sns.countplot(x=df['gender'],hue=df['Churn'])
plt.xlabel('Gender',fontweight='bold',fontsize=15)
plt.ylabel('Churn Count',fontweight='bold',fontsize=15)
plt.title('Gender vs Churn',fontweight='bold',fontsize=15)
```

# Find out how many female senior citizens there in the dataset

```python
fsc=df[(df['gender']=='Female') & (df['SeniorCitizen']==1)]
```

```python
fsc.head()
```

```python
fsc['gender'].count()
print('Number of female senior citizens there in the dataset =',fsc['gender'].count())
```

# Compare 'tenure' with 'Total Charges'

```python
plt.figure(figsize=(10,10))
plt.scatter(df['tenure'],df['TotalCharges'],s=10,marker='^')
plt.title('Tenure with Total Charges',fontweight='bold',fontsize=20)
plt.xlabel('Tenure',fontsize=15)
plt.ylabel('TotalCharges',fontsize=15)
```

# Find out which contract is preferred by the senior citizen.

```python
cz=df[df['SeniorCitizen']==1]
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x=cz['Contract'],hue=cz['Contract'])
plt.title('contract preferred by the senior citizen',fontweight='bold',fontsize=20)
plt.xlabel('Contract Type',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.grid()
plt.show()
```
